1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
newRestaurant = {
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}
```

```python
# Insert the new restaurant into the collection
db.establishments.insert_one(newRestaurant)
```

```python
# Check that the new restaurant was inserted
print('New restaurant inserted:')
query = {'BusinessName': 'Penang Flavours'}
checkRestaurantInserted = establishments.find(query)

for restaurant in checkRestaurantInserted:
    print(restaurant)
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query_BusinessTypeID = {'BusinessType': 'Restaurant/Cafe/Canteen'}

# Define the projection to return only the BusinessTypeID and BusinessType fields
query_Fields = {'_id': 0, 'BusinessTypeID': 1, 'BusinessType': 1}

# Execute the query and projection
results_BusinessTypeID = establishments.find(query_BusinessTypeID, query_Fields)

# Use for loop to interate over the results_BusinessTypeID and print each document
for businessType in results_BusinessTypeID:
    print(businessType)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
query_PenangFlavours = {'BusinessName': 'Penang Flavours'}

# Update the Penang Flavours record's BusinessTypeID to '1'
update_PenangFlavours = {'$set': {'BusinessTypeID': '1'}}

# Execute the update
result = establishments.update_one(query_PenangFlavours, update_PenangFlavours)
```

```python
# Confirm that the new restaurant was updated
if result.matched_count > 0:
    print('Update successful.')
    if result.modified_count > 0:
        print('The document was modified.') 
        # Fetch and print the updated document
        updated_PenangFlavours = establishments.find_one({'BusinessName': 'Penang Flavours'})
        print('Updated document:', updated_PenangFlavours)
    else:
        print('The document was already up to date.')
else:
    print('No matching document found.')
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for latitude and longitude
# Call the .update_many on the establishments collection and set up a query parameter to update with an 
# aggregation pipeline to convert the latitude and longitude values to Double
establishments.update_many({}, [{'$set': {'geocode.latitude': {'$toDouble': '$geocode.latitude'},
                                          'geocode.longitude': {'$toDouble': '$geocode.longitude'}
                                         }
                                }
                               ]
                          )
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
# Call the .update_many on the establishments collection and set up a query parameter to update with an 
# aggregation pipeline to convert the RatingValue values to Integer
establishments.update_many({}, [{'$set': {'RatingValue': {'$toInt': '$RatingValue'}}}])
```

```python
# Check that the coordinates and rating value are now numbers
# Fetch a sample of documents
sample_docs = establishments.find({}, {'geocode': 1, 'RatingValue': 1}).limit(5)

# Iterate over sample_docs and check data types
for doc in sample_docs:
    latitude = doc.get('geocode', {}).get('latitude')
    longitude = doc.get('geocode', {}).get('longitude')
    rating_value = doc.get('RatingValue')

    print(f"Document ID: {doc.get('_id')}")
    print(f"Latitude type: {type(latitude).__name__} - Value: {latitude}")
    print(f"Longitude type: {type(longitude).__name__} - Value: {longitude}")
    print(f"RatingValue type: {type(rating_value).__name__} - Value: {rating_value}")
    print("---")
```
